### Question 1: What is overfitting?

**Answer:** Overfitting occurs when a model learns the training data too well, capturing noise and specific patterns that don't generalize to new, unseen data.

### Question 2: What is underfitting?

**Answer:** Underfitting occurs when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and new data.

### Question 3: What is a bias-variance trade-off?

**Answer:** It's the conflict in simultaneously minimizing two sources of error that prevent supervised learning algorithms from generalizing beyond their training data: bias (error from erroneous assumptions) and variance (error from sensitivity to small fluctuations in the training set).

### Question 4: What is regularization?

**Answer:** Regularization is a technique used to prevent overfitting by adding a penalty term to the loss function, discouraging overly complex models.

### Question 5: Explain gradient descent.

**Answer:** Gradient descent is an optimization algorithm used to minimize a function by iteratively moving in the direction of steepest descent (negative of the gradient).

### Question 6: What is cross-validation?

**Answer:** Cross-validation is a technique to evaluate a model's performance on unseen data by partitioning the data into subsets, training on some, and testing on others, iteratively.

### Question 7: What is the difference between classification and regression?

**Answer:** Classification predicts a discrete label (e.g., cat/dog), while regression predicts a continuous value (e.g., house price).

### Question 8: What is a confusion matrix?

**Answer:** A confusion matrix is a table used to evaluate the performance of a classification model by summarizing correctly and incorrectly classified instances.

### Question 9: What is feature engineering?

**Answer:** Feature engineering is the process of creating new features or modifying existing ones from raw data to improve the performance of machine learning models.

### Question 10: What is an activation function?

**Answer:** An activation function in a neural network introduces non-linearity, allowing the network to learn complex patterns and make decisions.
