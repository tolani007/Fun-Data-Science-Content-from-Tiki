# Task
Analyze the "customer_info.csv" dataset to identify customers for a marketing campaign. The campaign targets customers with 100-999 employees, in the 'Education' or 'Food Services' industries, and with a total spend on mobility and internet services less than $5000. Calculate the total number of eligible customers, cross-tabulate their distribution by 'INDUSTRY' and 'STATE', visualize this distribution, and export the 'CUSTOMER_ID' of these customers to a new CSV file named 'customerids.csv'. Finally, summarize the key findings from this analysis.

## Identify Campaign Customers

### Subtask:
Create a boolean column named 'INCLUDE' that is True for customers who meet all campaign criteria: 100 <= EMP <= 999, INDUSTRY is 'Education' or 'Food Services', and TOTAL_SPEND < $5000.


**Reasoning**:
To create the 'INCLUDE' column based on the specified criteria, I will apply a series of boolean conditions to the `customer_file_object` DataFrame, combining them with logical operators.



```python
customer_file_object['INCLUDE'] = (
    (customer_file_object['EMP'] >= 100) &
    (customer_file_object['EMP'] <= 999) &
    (customer_file_object['INDUSTRY'].isin(['Education', 'Food Services'])) &
    (customer_file_object['TOTALSPEND'] < 5000)
)

print(customer_file_object.head())
```

## Count Campaign Customers

### Subtask:
Calculate and display the total number of customers included in the campaign by counting the True values in the 'INCLUDE' column.


**Reasoning**:
To calculate the total number of customers included in the campaign, I will access the 'INCLUDE' boolean column from the 'customer_file_object' DataFrame and sum its values. Since True is treated as 1 and False as 0, the sum will directly give the count of True values.



```python
num_eligible_customers = customer_file_object['INCLUDE'].sum()
print(f"Total number of customers included in the campaign: {num_eligible_customers}")
```

## Cross-Tabulate Campaign Customer Distribution

### Subtask:
Use the `pd.crosstab` function to determine how many customers in each industry and state are included in the campaign. Display the resulting cross-tabulation.

#### Instructions
1. Filter the `customer_file_object` DataFrame to include only customers where the 'INCLUDE' column is `True`.
2. Create a cross-tabulation of the 'INDUSTRY' and 'STATE' columns for these filtered customers using `pd.crosstab`.
3. Display the cross-tabulation.

**Reasoning**:
To perform the cross-tabulation as described in the subtask, I will first filter the DataFrame to only include customers marked as 'INCLUDE', and then apply `pd.crosstab` to the 'INDUSTRY' and 'STATE' columns of this filtered DataFrame.



```python
campaign_customers = customer_file_object[customer_file_object['INCLUDE'] == True]
crosstab_result = pd.crosstab(campaign_customers['INDUSTRY'], campaign_customers['STATE'])
print(crosstab_result)
```

## Visualize Campaign Customer Distribution

### Subtask:
Visualize the distribution of campaign customers by 'INDUSTRY' and 'STATE' using a stacked bar chart. This will provide a clear visual representation of how many eligible customers are in each industry within each state.

#### Instructions
1. Use the `crosstab_result` DataFrame to create a stacked bar plot.
2. Ensure the plot has appropriate labels for the x-axis, y-axis, and a title.
3. Display the plot.

**Reasoning**:
To visualize the distribution of campaign customers by industry and state, I will create a stacked bar plot using `crosstab_result`. This requires importing `matplotlib.pyplot` for plotting functions and will ensure the plot has appropriate labels and a title for clarity.



```python
import matplotlib.pyplot as plt

# Plotting the stacked bar chart
crosstab_result.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Campaign Customers Distribution by Industry and State')
plt.xlabel('Industry')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.legend(title='State')
plt.tight_layout()
plt.show()
```

## Export Campaign Customer IDs

### Subtask:
Create a file called `customerids.csv` that contains only a list of `CUSTOMER_ID`s for customers included in the campaign.

#### Instructions

2. From these filtered rows, select the 'CUSTOMER_ID' column.

```python
campaign_customer_ids = customer_file_object[customer_file_object['INCLUDE'] == True]['CUSTOMER_ID']
campaign_customer_ids.to_csv('customerids.csv', index=False)
print("customerids.csv created successfully.")
```
